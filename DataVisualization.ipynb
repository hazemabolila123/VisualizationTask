```python
#import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the dataset
mpg = pd.read_csv("E:/hazem/Technology/Data science/CAT/5-Data Visulization/task/auto-mpg.csv")
mpg.head(n=10)
```

```python
#Getting inforamtion abuot the DataFrame
mpg.info()
```

```python
#Discover model year Column
g =sns.countplot(x="model year",data=mpg)
g.set_title("Car Model Year", color="r")              
plt.show()
```

```python
#Discover origin Column
g =sns.countplot(x="origin",data=mpg)
g.set_title("Origin", color="r") 
plt.show()
```

```python
#Relation between weight and acceleration per origin
g =sns.relplot(x="weight",y="acceleration",data=mpg,kind="scatter" , col ="origin" )
g.fig.suptitle("How weight affects acceleration",y=1.1, color="r")
plt.show()
```

```python
#Relation between horsepower and acceleration per origin
g= sns.relplot(x="horsepower",y="acceleration",data=mpg,kind="scatter" ,col ="origin",hue="cylinders")
sns.set_context("talk")
plt.xticks(rotation=90)
g.fig.suptitle("How horsepower affects acceleration",y=1.1, color="r")
plt.show()
```

```python
#Relation between horsepower and mpg per origin
g= sns.relplot(x="horsepower",y="mpg",data=mpg,kind="scatter" ,col ="origin",hue="cylinders")
sns.set_context("talk")
plt.xticks(rotation=90)
g.fig.suptitle("How horsepower affects mpg",y=1.1, color="r")
plt.show()
```

```python
#How the displacement of the engine affects mpg
g = sns.relplot(x="displacement",y="mpg", data=mpg,kind="scatter",col="origin" )
g.fig.suptitle("How displacement affects mpg per each origin" , y=1.1,x=.45 , color="r")
plt.show()
```

```python
#Relation between displacement and acceleration
g=sns.relplot(x="displacement",y="acceleration",data=mpg,kind="scatter",col="origin")
g.fig.suptitle("Relation between displacement and acceleration per origin", color="r" ,y=1.1,x=.45)
```

```python
#Relation between the increaing of no of cylinders and mpg
g=sns.catplot(x="cylinders",y="mpg",data=mpg,kind="box", col="origin")
g.fig.suptitle("How numbre of cylinders affects mpg",y=1.1,x=.45,color="r")
plt.show()
```

```python
#Acceleration Per Origin
g=sns.catplot(x="origin" , y="acceleration" ,data=mpg,kind="box")
g.fig.suptitle("Displacement Per Origin",y=1.1,x=.6,color="r")
plt.show()
```

```python
#Displacement Per Origin
g=sns.catplot(x="origin" , y="displacement" ,data=mpg,kind="box")
g.fig.suptitle("Displacement Per Origin",y=1.1,x=.6,color="r")
plt.show()
```

```python
#The average acceleration per each model year
g = sns.catplot(x="model year",y="acceleration",data=mpg,kind="bar" ,col= 'origin')
g.fig.suptitle("The average acceleration per each model year",y=1.1,x=.45,color="r")
plt.show()
```

```python
#The average displacement per each model year
g = sns.catplot(x="model year",y="displacement",data=mpg,kind="bar" , col="origin")
g.fig.suptitle("The average displacement per each model year",y=1.1,x=.45,color="r")
plt.show()
```

```python
#Mean of weights per years
g=sns.catplot(x="model year",y="weight",data=mpg,kind="point",join=False)
plt.xticks(rotation=90)
g.fig.suptitle(" Average weight through Years",y=1.1,x=.6, color="r")
plt.show()
```

```python
#Acceleration per years
g = sns.relplot(x='model year', y="acceleration" , data=mpg,kind="line")
g.fig.suptitle("acceleartion through years" , y=1.1 , x=.6 , color="r")
plt.show()
```

```python
#MPG per years
g = sns.relplot(x='model year', y="mpg" , data=mpg,kind="line" , ci=None)
g.fig.suptitle("mpg through years" , y=1.1 , x=.6, color="r")
plt.show()
```
